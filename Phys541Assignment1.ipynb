## Problem 2 Part C

```python
# x = K_BT/ℏw = 1/βℏw
### C_cl = 3NK_B  --> Ycl = C_cl/NK_B=3


x=  np.linspace(0,3,1000,endpoint=True) ; x[0]=1e-100

y_cl = 3
y_qu = 3*(1/x)**2 *np.exp(1/x)/( np.exp(1/x)-1)**2

fig = plt.figure(figsize=[6,5])
plt.plot(x,[y_cl]*len(x),color='darkred',label='Classical')
plt.plot(x,y_qu, color='darkblue' ,label='Quantum')

plt.ylim(0,4)

plt.title('Specific Heat for N 3D Simple Harmonic Oscillators ')
plt.ylabel(r'$\frac{C_v}{N K_B}$', rotation=0,labelpad=20,fontsize=15)
plt.xlabel(r'$\frac{K_B T}{\hbar \omega}$',labelpad=10,fontsize=15)
plt.legend() 
plt.show()
```

## Problem 3

```python
# x = T/T_D
# y = Cv/(N K_B)

y_debye=[]
x =  np.linspace(0,3,1000,endpoint=True) 

def integrand(x):
    return x**4*np.exp(x)/(np.exp(x)-1)**2

for i in x: 
    integral=quad(integrand, 0, 1/i)[0] ; y= 9*i**3*integral ; y_debye.append(y)

fig = plt.figure(figsize=[6,5])
plt.plot(x,y_debye, color='darkgreen' ,label='Debye')
plt.plot(x,y_qu, color='darkblue' ,label='Quantum')
plt.plot(x,[y_cl]*len(x),color='darkred',label='Classical')

plt.ylim(0,4)

plt.title('Specific Heat for N 3D Simple Harmonic Oscillators ')
plt.ylabel(r'$\frac{C_v}{N K_B}$', rotation=0,labelpad=20,fontsize=15)
plt.xlabel(r'$\frac{T}{T_D}$',labelpad=10,fontsize=15)

plt.legend() 
plt.show()
```

```python
fig = plt.figure(figsize=[6,5])

plt.loglog(x,y_debye, color='darkgreen' ,label='Debye')
plt.loglog(x,y_qu, color='darkblue' ,label='Quantum')
plt.loglog(x,[y_cl]*len(x),color='darkred',label='Classical')
plt.title('LogLog Specific Heat for N 3D Simple Harmonic Oscillators ')
plt.ylabel(r'$\frac{C_v}{N K_B}$', rotation=0,labelpad=20,fontsize=15)
plt.xlabel(r'$\frac{T}{T_D}$',labelpad=10,fontsize=15)

plt.legend() 
plt.show()
```

### Riemann zeta function

```python
def Riemann_zeta_function(s):
    zeta=0
    for n in range(1,10000000):
        zeta+=1/n**s
    return print('ζ({})={}'.format(s,zeta))
```
